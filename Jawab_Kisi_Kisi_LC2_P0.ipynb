#####1. Berapa banyak label unique dari bike_type?

````python
# working area
df_1 = client.query("""
SELECT count(distinct bike_type) as type_count FROM `ftds-hacktiv8-project.practice.bike_rent`""").to_dataframe()

df_1
````

#####2. Apa saja nama label unique dari tipe sepeda?

````python
# working area
df_1 = client.query("""
SELECT distinct bike_type FROM `ftds-hacktiv8-project.practice.bike_rent`""").to_dataframe()

df_1
````

#####3. Berapa banyak total peminjaman sepeda pada tahun 2016 dan 2017?

````python
# working area
df_1 = client.query("""
SELECT count(bike_id) as bikes FROM `ftds-hacktiv8-project.practice.bike_rent`
where extract(year from start_time) =2016 or extract(year from start_time)=2017""").to_dataframe()

df_1
````

#####4. Dimana station yang paling banyak dikunjungi sebagai titik keberangkatan pada tahun 2016?

````python
# working area
df_1 = client.query("""
  SELECT start_station_name, count(start_station_name) as maxcount FROM `ftds-hacktiv8-project.practice.bike_rent`
  where extract(year from start_time) =2016
  group by start_station_name
  order by maxcount desc
  limit 1
""").to_dataframe()

df_1
````

#####5. Berapa nilai minimum, maximum, dan rata-rata dari durasi perjalanan pada keseluruhan data?

````python
# working area
df_1 = client.query("""
  SELECT max(duration_minutes) as max_value, min(duration_minutes) as min_value, avg(duration_minutes) as average FROM `ftds-hacktiv8-project.practice.bike_rent`
""").to_dataframe()

df_1
````

#####6. Berapa rata-rata durasi perjalanan dari subscriber_type `Student Membership` dan `Local31`?

````python
# working area
df_1 = client.query("""
  SELECT subscriber_type, avg(duration_minutes) as average FROM `ftds-hacktiv8-project.practice.bike_rent`
  where subscriber_type = 'Student Membership' or subscriber_type = 'Local31'
  group by subscriber_type
""").to_dataframe()

df_1
````

#####7. Tampilkan entry data yang memiliki `duration_time` diatas 4.25 jam. (*Gunakan subquery untuk menjawab pertanyaan ini dengan meng-convert duration_time dari menit ke hour*).

````python
# working area
df_1 = client.query("""
  SELECT trip_id, hours FROM (
    select duration_minutes/60 as hours
    FROM `ftds-hacktiv8-project.practice.bike_rent`
    where duration_minutes/60 > 4.25), `ftds-hacktiv8-project.practice.bike_rent`
""").to_dataframe()

df_1
````

#####2. Buat kolom baru bernama `quarter`.
  
  Lakukan ekstraksi waktu (quarter dalam tahun) berdasarkan kolom `start_time` dengan ketentuan berikut:

<div align="center">

  |Keterangan Waktu|Bulan|
  |--|--|
  |Q1|Januari, Februari, Maret|
  |Q2|April, Mei, Juni|
  |Q3|Juli, Agustus, September|
  |Q4|Oktober, November, Desember|

</div>

>**Notes:**
>- Dalam menjawab soal Pandas no. 2, **tidak boleh** menggunakan ekstraksi waktu `Series.dt.quarter`.
- Lakukan `looping` dan `conditional` dengan memanfaatkan ekstraksi bulan `Series.dt.month` terhadap kolom `start_time`, pastikan tipe datanya ialah `datetime` agar dapat dilakukan ekstraksi.
- Simpan hasil looping ke dalam suatu variabel, dan buat kolom baru berisikan variabel sebelumnya.

````python
# working area
df['month']=df['start_time'].dt.month
df['quarter']= None #Inisialisasi kolom kosong untuk memasukan data quarter

for index, row in df.iterrows(): #.iterrows() digunakan agar bisa mendapatkan index
  if row['month'] == 1 or row['month'] == 2 or row['month'] == 3:
    df.at[index, 'quarter']='Q1'
  elif row['month'] == 4 or row['month'] == 5 or row['month'] == 6:
    df.at[index, 'quarter']='Q2'
  elif row['month'] == 7 or row['month'] == 8 or row['month'] == 9:
    df.at[index, 'quarter']='Q3'
  else:
    df.at[index, 'quarter']='Q4'
df[['month','quarter']]
````

#####3. Tampilkan data perjalanan dari tipe member `Local31` dengan durasi perjalanan diatas 1 jam.
*(Jika tidak ada entry yang memenuhi kondisi ini tidak apa, tuliskan kode pandasnya dengan lengkap saja)*

````python
# working area
df[(df['subscriber_type']=='Local31') & (df['duration_minutes']>60)]
````

#####4. Berapa rata-rata durasi perjalanan dari tipe pelanggan `Walk Up` dengan tipe sepeda `classic` pada bulan `Maret` dari data yang diambil?
*(Jika tidak ada entry yang memenuhi kondisi ini tidak apa, tuliskan kode pandasnya dengan lengkap saja)*

````python
# working area
df.query("month == 3 and subscriber_type == 'Walk Up' and bike_type == 'classic'")['duration_minutes'].mean()
````
